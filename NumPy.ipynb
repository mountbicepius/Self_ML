```python
import xlsxwriter

# Create an new Excel file and add a worksheet.
workbook = xlsxwriter.Workbook('Pandas10.xlsx')
worksheet = workbook.add_worksheet()

# Widen the first column to make the text clearer.
worksheet.set_column('A:A', 30)

# Add a bold format to use to highlight cells.
bold = workbook.add_format({'bold': True})

# Write some simple text.
worksheet.write('A1', 'Dept of CA')
# Text with formatting.
worksheet.write('A2', 'RCCIIT', bold)

worksheet.write('A3', 'Roll', bold)
worksheet.write('B3', 'Name', bold)

# Write some students information with row/column notation.
worksheet.write(4, 0, 'MCA2021/001')
worksheet.write(4, 1, 'Abcd')
worksheet.write(5, 0, 'MCA2021/002')
worksheet.write(5, 1, 'Efgh')


workbook.close()
```

```python
# import module sys to get the type of exception
import sys

randomList = [5,10,'a', 0, 2,88,100,'abc',66,0,44]

for entry in randomList:
    try:
        print("The entry is", entry)
        r = 1/int(entry)
        #break
    except:
        print("Oops!", sys.exc_info()[0], "occurred.")
        print("Next entry.")
        print()
print("The reciprocal of", entry, "is", r)
```

```python
#Since every exception in Python inherits from the base Exception class, 
# we can also perform the above task in the following way:
# import module sys to get the type of exception
import sys

randomList = [5,10,'a', 0, 2]

for entry in randomList:
    try:
        print("The entry is", entry)
        r = 1/int(entry)
        #break
    except Exception as e:
        print("Oops!", e.__class__, "occurred.")
        print("Next entry.")
        print()
print("The reciprocal of", entry, "is", r)
```

```python
# Raising Exceptions in Python
# In Python programming, exceptions are raised when errors occur at runtime.
# We can also manually raise exceptions using the raise keyword.

try:
    a = int(input("Enter a positive integer: "))
    if a <= 0:
        raise ValueError("That is not a positive number!")
except ValueError as ve:
    print(ve)   
```

```python
# Python try with else clause
# In some situations, you might want to run a certain block of code if the code block inside try ran without any errors.
# For these cases, you can use the optional else keyword with the try statement.

# program to print the reciprocal of even numbers

try:
    num = int(input("Enter a number: "))
    assert num % 2 == 0
except:
    print("Not an even number!")
else:
    reciprocal = 1/num
    print(reciprocal)
```
